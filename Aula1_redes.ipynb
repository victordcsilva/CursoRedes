# Básico de Redes de Computadores

Podemos classificar os computadores em três tipos:
- __Desktop ou de mesa__
- __Portáteis__:  Notebooks e tablets
- __Propósito específicos__: Smartphones, videogames, smart TVs, impressoras

## O que é uma rede de computadores?

- Dois ou mais computadores conectados entre si por intermédio de algum meio que permita a comunicação entre estes computadores.
- Não basta só a conexão, é necessário haver a comunicação. 
- É mais comum a comunicação por intermédio de algum dispositivo que facilita o processo, como roteadores. 

__Roteador__

- Gerencia o tráfego de dados entre os computadores e dispositivos. 
- Existem diversos tipos de roteadores com capacidades limitadas até com capacidades de interligar centenas de computadores. 

## Objetivo de uma rede de computadores

- Troca de dados entre computadores.
- Pequenas empresas compartilham recursos (impressoras, scanners) e dados (arquivos)

## Classificação das redes

- Podemos classificar as redes de computadores quanto:
    - Extensão geográfica;
    - Meio de transmissão;
    - Arquitetura; e 
    - Topologia. 
    
- __Extensão Geográfica:__
    - __LANs (Local Area Networks - Redes Locais)__ Interligam computadores presentes dentro de um mesmo espaço físico, empresa, escola, casa.  
    - __MANs (Metropolitan Area Network - Redes Metropolitana)__ Conecta diversas LANs dentro de algumas dezenas de quilômetros. 
    - __WANs (Wide Area Networks - Redes de Longa Distância)__ Conecta diversas LANs em uma área maior, como um país ou continente. 
    - __PANs (Personal Area Networks - Redes Pessoais)__ Conecta dispositivos dentro de uma distância bastante limitada.
    - __Wirelesse (Sem Fio)__ Versão sem fio das categorias acima. 
    
- __Redes de Telefonia Celular__
    - Um celular está conectado a uma WAN sem fio ou LAN sem fio.
    - Sinal de celular vem da chamada rede móvel (são as operadoras). 
    
- __Redes Cabeadas__ 
    - Dois ou mais computadores se comunicam usando algum tipo de cabo. 
    - Mesmo possivel a conexão direta entre dois computadores, o ideal é usar um roteador e ligar os computadores no roteador. 
    - Roteadores podem se conectar a outros roteadores, podendo se comunicar com computadores que estão conectados em outros roteadores. 
    - A tecnologia mais popular para redes cabeadas é a Ethernet. 
    - A vantagem da rede cabeada a rede sem fio é a velocidade. 
    - A desvantagem é a necessidade de ter um cabo ligando um equipamento ao outro. 
    
    ![fig_cabo_ethernet](https://pplware.sapo.pt/wp-content/uploads/2019/11/ethernet_3-720x409.jpg)
    Figura: Cabo ethernet sendo conectado a um roteador. 
    
- __Redes Sem Fio__
    - Dois ou mais computadores se comunicam usando um sinal de rádio.
    - É necessario um equipamento atuar intermediando a comunicação entre os dois ou mais equipamentos. 
    - Muitos roteadores fornecem conexão por rede cabeada e rede sem fio.
    - Uma desvantagem é o alcance, dependendo da forma da casa, quantidade de paredes e frequência do sinal de rádio, o sinal pode chegar no equipamento muito fraco, gerando erros na comunicação.
    - Outra desvantagem é a interferência, equipamentos eletronicos ou até outros equipamentos de rede sem fio podem prejudicar o sinal.

## World Wide Web e Internet 

- WWW é uma aplicação ou serviço que funciona na Internet.
- Internet é uma rede mundial de computadores, é uma estrutura física.
- Internet: Estrutura de redes de alcance mundial, que permite a interconexão de diversas outras redes de computadores espalhadas pelo mundo. 
- Uma rede físicamente não é muito útil. O importante é o que é possível fazer com esta rede, ou seja, suas aplicações e serviços. 
- A internet oferece diversos serviços (email, tranferência de arquivos).
- O mais famoso é a Web (World Wide Web) 
- Web: Conjunto de tecnologias que permite transferir um tipo de documento especial, um arquivo digital de texto puro, o HTML, atráves de um protocolo de comunicação chamado HTTP. 
    - Protocolo é um conjunto de regras para funcionamento de algo. 
- Esses arquivos HTML são visualizados atráves de programas de computadores chamados de Navegadores Web (Google Chrome, Opera, Firefox...)
- Os navegadores entendem as informações presentes nos HTTP e transformam em páginas web. 
- __Servidores Web__ Computadores que tem como função oferecer os dados necessários para que possamos navegar na internet. 
    - Sua principal função é servir dados para outros computadores. 
    - Os computadores que recebem os dados dos servidores são chamados de clientes. 
- __Provedores de Internet__ Empresas que oferecem o acesso a internet para os usuários comuns.
    - Controlam a velocidade da internet.
    - Controlam a quantidade de dados que você transfere dos servidores para o seu computador. 
    
##  TCP/IP

- Conjunto de protocolos (regras) de comunicação entre computadores. 
- Para um computador se conectar a internet, ele precisa de uma conexão com a internet e entender a língua da internet. 
- Ao navegar na internet estamos recebendo pacotes de dados codificados para o formato de comunicação TCP/IP. 
- Cada computador dentro de uma rede TCP/IP possui um identificador único, chamado endereço IP (por exemplo: 192.168.0.1). 
- Em uma rede local, esse endereço é oferecido pelo modem, pelo roteador ou automaticamente definido pelo sistema operacional. 
- Servidores e todos os computadores na Internet possuem seu próprio endereço IP.
- Redes locais, mesmo sem acesso a internet, pode ter endereços IP.
